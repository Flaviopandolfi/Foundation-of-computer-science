Nei ranking sono presenti alcune università il cui posizionamento non è indicato più con una posizione precisa, bensì come una possibile posizione all'interno di un range. Non avendo modo di determinare la sua posizione precisa si è scelto di calcolare il valore medio del range e utilizzarlo come valore del ranking. Questa scelta è dettata dal fatto che il valore medio è quello che meglio rappresenta l'intervallo nel complesso ed è una stima equilibrata (nè ottimistica nè pessimistica) della posizione reale nel ranking.

Sì è reso necessario fare un doppio extract. Se ne avessimo fatto uno solo la colonna di destra avrebbe contenuto valori NaN ogni volta che la posizione nel ranking non fosse stata espressa in forma intervallare. Con il codice precedente, invece, in Times2 vengono posti tutti i numeri singoli e tutti i numeri seguiti da un trattino, mentre in Times3 vengono messi tutti i numeri che terminano la stringa. In questo modo, in presenza di un range, in Times2 avremo i numeri dell'estremo sinistro (fino al -) e in Times3 in numeri dell'estremo destro (dopo il -). Se ci fosse un numero solo avremmo lo stesso numero sia in Times2 che in Times3.

Per ottenere la differenza massima tra i tre ranking per ciascuna università procedo nel seguente modo: 
   1. CREO UN DATA-FRAME Ti_Sh_Cwr CONTENENTE TUTTE LE UNIVERSITA' E I VALORI DI CIASCUN RANKING (SOLO I PIU' RECENTI).
   2. CALCOLO LA DIFFERENZA TRA I RANKING.
   3. DEFINISCO UNA FUNZIONE CHE ESTRAGGA PER OGNI UNIVERSITA' LA DIFFERENZA MAGGIORE IN VALORE ASSOLUTO.

```python
MRS=shanghai_new[['university_name', 'world_rank']]
```

```python
Ti_Sh=pd.merge(MRT, MRS, on='university_name')
Ti_Sh
#times + shanghai
```

Avendo fatto il merge usando university_name come chiave ora ho una tabella con i 3 diversi ranking associati a ciascuna università

```python
#2
Ti_Sh_Cwr['ti_sh_rank_diff']=Ti_Sh_Cwr['world_rank_x']-Ti_Sh_Cwr['world_rank_y']
Ti_Sh_Cwr['ti_cwr_rank_diff']=Ti_Sh_Cwr['world_rank_x']-Ti_Sh_Cwr['world_rank']
Ti_Sh_Cwr['sh_cwr_rank_diff']=Ti_Sh_Cwr['world_rank_y']-Ti_Sh_Cwr['world_rank']
```

```python
#3
def max_difference(riga):
    ti_sh=abs(riga['ti_sh_rank_diff'])
    sh_cw=abs(riga['sh_cwr_rank_diff'])
    ti_cw=abs(riga['ti_cwr_rank_diff'])
    return max(ti_sh, sh_cw, ti_cw)
```

```python
Ti_Sh_Cwr['max_difference']=Ti_Sh_Cwr.apply(max_difference, axis=1)
Ti_Sh_Cwr
```

# 5. Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.

Le celle [52] e [54] creano due colonne 'female_ratio' e 'male_ratio'contenenti solo i numeri a sinstra (la prima) e a destra (la seconda) dei due punti in ciascuna cella della colonna 'female_male_ratio'. E' simile al procedimento di extract

```python
pd.options.mode.chained_assignment = None
most_recent_time['male_ratio'] = most_recent_time['female_male_ratio'].str.replace(':', '')
most_recent_time['male_ratio'] = most_recent_time['male_ratio'].str.replace('^\d\d', '')
most_recent_time['male_ratio'] = pd.to_numeric(most_recent_time['male_ratio'], errors='coerce')
```

```python
most_recent_time['num_students'] = most_recent_time['num_students'].str.replace(',', '')
most_recent_time['num_students'] = pd.to_numeric(most_recent_time['num_students'], errors='coerce')
```

La cella [53] converte in numero il contenuto della colonna 'num_students', laddove la conversione non fosse possibile verrebbe inserito un valore NaN. Per poter convertire efficacemente è stato necessario eliminare le ',' utilizzate dagli anglosassoni per separare le migliaia dalle centinaia (ogni 3 cifre).

```python
most_recent_time['female_ratio'] = most_recent_time['female_male_ratio'].str.replace(':', '')
most_recent_time['female_ratio'] = most_recent_time['female_ratio'].str.replace('\d\d$', '')
most_recent_time['female_ratio']
most_recent_time['female_ratio'] = pd.to_numeric(most_recent_time['female_ratio'], errors='coerce')
```

```python
most_recent_time['number_female'] = round(most_recent_time['female_ratio']*most_recent_time['num_students']/100)
most_recent_time['number_female']
```

Poichè la somma dei due elementi del female_male_ratio corrispondeva sempre a 100 si è supposto che indicassero la % di studentesse femmine e di studenti maschi. Per questo il numero di studentesse è stato calcolato come percentuale del totale di studenti. Da qui la formula 'number_female'= 'female_ratio' * 'num_students'/100. Il valore ottenuto è stato arrotondato perchè il numero di persone dev'essere per forza intero (arrotonda per difetto se parte decimale <= 0.5).
Allo stesso modo si calcolerà 'number_male'. Eventualmente si potrebbe fare anche la semplice differenza tra 'num_students' e 'number_female'

```python
most_recent_time['number_male'] = round(most_recent_time['male_ratio']*most_recent_time['num_students']/100)
most_recent_time['number_male']
```

```python
pd.options.mode.chained_assignment = "warn"
```
